```python
df.loc[y==1,'Customer service calls'].hist();   # churn clients
df.loc[y==0,'Customer service calls'].hist();
plt.xlabel('calls')
plt.ylabel('frequency')
plt.title('Customer service calls')
```

```python
df.loc[y==1,'Customer service calls'].hist(alpha=.5,label='Churn');   # churn clients
df.loc[y==0,'Customer service calls'].hist(alpha=.5, label='loyal');
plt.xlabel('calls')
plt.ylabel('frequency')
plt.title('Customer service calls')
```

```python
df.loc[y==1,'Customer service calls'].hist(alpha=.5,label='Churn', normed=True);   # churn clients
df.loc[y==0,'Customer service calls'].hist(alpha=.5, label='loyal', normed=True);
plt.xlabel('calls')
plt.ylabel('frequency')
plt.legend()
plt.title('Customer service calls')
```
